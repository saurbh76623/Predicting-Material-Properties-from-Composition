```python
X = df[['Atomic Radius(Atomic unit)', 'Expansion Coefficient', 'Electron Affinity(-ive)', 'Conductivity', 'Yield Strength']]  # Input features
y = df['Outcome']  # Output/target variable
```

```python
X  #input columns
```

```python
y # output cols
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# imputer = SimpleImputer(strategy='most_frequent')
# y = imputer.fit_transform(y)

if y.isnull().any():
    y = y.fillna(y.mode()[0])
```

# scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_predict)*100
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_predict)*100
```

```python
from sklearn.svm import SVC

model = SVC(kernel='rbf')
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_predict)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score(y_test,y_predict)*100
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

# Predict probabilities
y_pred_prob = model.predict(X_test)

y_pred = np.where(y_pred_prob > 0.5, 1, 0)

# Now you can calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

#Hyperparameters tuning

```python
!pip install scikeras

from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier  # Import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense

# Define your Keras model
def create_model(learning_rate=0.001): #added learning_rate parameter
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) #removed learning_rate
    return model

# Create a KerasClassifier wrapper
model = KerasClassifier(model=create_model, learning_rate=0.001, verbose=0) # wrap the model

param_grid = {
    'batch_size': [16, 32, 64],
    'epochs': [10, 20],
    'learning_rate': [0.001, 0.01] #added learning_rate in param_grid
}


grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_}")
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_distributions = {
    'batch_size': [16, 32, 64],
    'epochs': [10, 20],
    'learning_rate': [0.001, 0.01]
}

random_search = RandomizedSearchCV(estimator=model, param_distributions=param_distributions, n_iter=10, scoring='accuracy', n_jobs=-1)
random_search.fit(X_train, y_train)

print(f"Best parameters: {random_search.best_params_}")
print(f"Best cross-validation score: {random_search.best_score_}")
```
